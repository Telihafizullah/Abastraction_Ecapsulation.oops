```python
# Q1. What is Abstraction in OOps? Explain with an example. 
#  Abstraction in Object-Oriented Programming (OOP) is the process of hiding a program's underlying implementation and details. 
# It's a core concept that simplifies things by focusing on relevant features and hiding unnecessary ones.
#  Abstraction can help manage complexity by splitting it up into smaller parts

# Q2. Differentiate between Abstraction and Encapsulation. Explain with an example. 
#  
# Abstraction and Encapsulation are two fundamental concepts in Object-Oriented Programming (OOP).
#  While they are often mentioned together, they serve different purposes: 


#Abastraction
# Focus:
# Hiding the implementation details and showing only the essential features of an object.
# Goal:
# Simplify interaction with complex systems by focusing on what an object does, not how it does it.
# Example:
# Consider a car. You interact with the steering wheel, pedals, and gear lever to drive it. 
# You don't need to know about the engine, transmission, or electronics that make it work. 

#Encapculation
# Focus: Bundling data and methods that operate on that data into a single unit, called a class.
# Goal: Protect the internal state of an object and control access to it.
# Example: In a car, the engine is encapsulated under the hood. You can't directly access or modify its parts.
# You interact with the car through its controlled interface (steering wheel, pedals, etc.)

# Q3. What is abc module in python? Why is it used?
# The abc module in Python stands for "Abstract Base Classes". It provides a way to define interfaces (contracts) that classes must adhere to. 
# This ensures consistency and structure in your code, especially when working with inheritance and polymorphism   
# By defining abstract methods in an abstract base class, you ensure that all subclasses implement those methods.
# This helps in creating a consistent interface across different classes and makes your code more robust and maintainable.

# Q4. How can we achieve data abstraction?   
# Abstract Classes:
# These are classes that cannot be instantiated directly. They serve as blueprints for other classes, defining a common interface
# that derived classes must implement.

# Abstract Methods:
# These are methods declared in an abstract class but without any implementation. Subclasses are required to provide concrete implementations
#  for these methods

# Q5. Can we create an instance of an abstract class? Explain your answer. 
# No, we cannot create an instance of an abstract class in Python
#Purpose of Abstract Classes:
# Abstract classes
# are meant to serve as blueprints or templates for other classes. They define a common interface that their subclasses must implement.

# Incomplete Implementation:
# Abstract classes can contain abstract methods, which are declared but not implemented. This means they lack complete functionality and cannot be 
# used directly.

# TypeError on Instantiation:
# Attempting to create an instance of an abstract class will raise a TypeError.



```
